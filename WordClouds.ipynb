```python
# Imports
import pandas as pd
import numpy as np
import nltk
import string
import re
import matplotlib.pyplot as plt
from pickle import dump, load
from PIL import Image


```

```python
# Read preprocessed csv
url = "https://uom-twitter-sentiment-analysis.s3.us-east-2.amazonaws.com/Lemmatize.csv"
lemmatized_df = pd.read_csv(url)
lemmatized_df.head()
```

```python
from wordcloud import WordCloud
WordCloud
cloud = WordCloud(background_color='white').generate(str(lemmatized_df['text']) )
fig= plt.figure(figsize=(14,14))
plt.imshow(cloud)
```

```python
negative_Tweets = lemmatized_df[lemmatized_df.target == 0]

negative_Tweets.head()

from wordcloud import WordCloud
wc = WordCloud(max_words=50)

cloud = WordCloud(background_color='white').generate(str(negative_Tweets ))
fig= plt.figure(figsize=(14,14))
plt.axis('off')
plt.imshow(cloud)
```

```python
positive_Tweets = lemmatized_df[lemmatized_df.target == 1]

positive_Tweets.head()

from wordcloud import WordCloud, STOPWORDS
wc = WordCloud(max_words=50)

cloud == WordCloud(background_color='white').generate(str(positive_Tweets ))
fig= plt.figure(figsize=(14,14))
plt.axis('off')
plt.imshow(cloud)
```

```python
thumbs_up_mask = np.array(Image.open('images/thumbs_up.jpg'))
thumbs_down_mask = np.array(Image.open('images/thumbs_down.jpg'))
```

```python
# TUWordCloud = WordCloud(mask =thumbs_up_mask,background_color='white').generate(str(positive_Tweets))
TUWordCloud = WordCloud(stopwords=STOPWORDS, background_color='gray', max_words=1000, mask=thumbs_up_mask,contour_color='#023075',contour_width=15,colormap='rainbow').generate(str(positive_Tweets))
plt.figure(figsize=(15,10))
plt.imshow(TUWordCloud, interpolation='bilinear')
plt.axis('off')
# plt.imshow(TUWordCloud)
```

```python
TDWordCloud = WordCloud(stopwords=STOPWORDS, background_color='gray', max_words=1000, mask=thumbs_down_mask,contour_color='#023075',contour_width=15,colormap='rainbow').generate(str(negative_Tweets))
plt.figure(figsize=(15,10))
plt.imshow(TDWordCloud, interpolation='bilinear')
plt.axis('off')
```
