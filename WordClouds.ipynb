```python
# Imports
import pandas as pd
import numpy as np
import nltk
import string
import re
import matplotlib.pyplot as plt
from pickle import dump, load

from PIL import Image
```

```python
# Read preprocessed csv
url = "https://uom-twitter-sentiment-analysis.s3.us-east-2.amazonaws.com/Lemmatize.csv"
lemmatized_df = pd.read_csv(url)
lemmatized_df.head()
```

```python
from wordcloud import WordCloud
# WordCloud
cloud = WordCloud().generate(str(lemmatized_df['text']) )
fig= plt.figure(figsize=(14,14))
plt.imshow(cloud)
```

```python
negative_Tweets = lemmatized_df[lemmatized_df.target == 0]

negative_Tweets.head()

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
wc = WordCloud(max_words=50)

cloud =wc.generate(str(negative_Tweets ))
fig= plt.figure(figsize=(14,14))
plt.imshow(cloud)
```

```python
positive_Tweets = lemmatized_df[lemmatized_df.target == 1]

positive_Tweets.head()

from wordcloud import WordCloud
wc = WordCloud(max_words=50)

cloud =wc.generate(str(positive_Tweets ))
fig= plt.figure(figsize=(14,14))
plt.imshow(cloud)
```

```python
thumbs_up_mask = np.array(Image.open('images/t3.png'))
thumbs_down_mask = np.array(Image.open('images/thumbs_down.png'))
```

```python
stopwords = set(STOPWORDS)
wordcloud = WordCloud(random_state = 1,background_color='white', max_words=1000, mask=thumbs_up_mask,\
                      contour_width=3, colormap='tab20b', collocations = False).generate(str(positive_Tweets))

plt.figure(figsize=[10,10])
plt.imshow(wordcloud)
plt.axis("off")
```

```python
stopwords = set(STOPWORDS)
wordcloud = WordCloud(random_state = 1,background_color='white', max_words=1000, mask=thumbs_down_mask,\
                      contour_width=3, colormap='tab20b', collocations = False).generate(str(negative_Tweets))
plt.figure(figsize=[10,10])
plt.imshow(wordcloud)
plt.axis("off")
```
